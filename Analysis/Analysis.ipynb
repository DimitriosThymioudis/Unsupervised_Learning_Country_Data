#### Correlations

```python
ut = np.triu(df.corr())
lt = np.tril(df.corr())

fig,ax = plt.subplots(nrows = 1, ncols = 2,figsize = (15,5))
plt.subplot(1,2,1)
sns.heatmap(df.corr(),cmap = 'YlOrBr',annot = True,cbar = 'True',mask = ut);
plt.title('Correlation Matrix : Upper Triangular Format');

plt.subplot(1,2,2)
sns.heatmap(df.corr(),cmap = 'YlOrBr',annot = True,cbar = 'True',mask = lt);
plt.title('Correlation Matrix : Lower Triangular Format');
```

#### Country vs Numerical Features

```python
top_10 = pd.DataFrame()
for column in df.columns:
    top_10 = df_original.sort_values(column,ascending=False)[['country',column]].head(5)
    plt.figure(figsize = (7,3))
    plt.title('Top 5 countries in : '+ column)
    sns.barplot(data=top_10, x="country", y=column);
```

```python
bot_10 = pd.DataFrame()
for column in df.columns:
    bot_10 = df_original.sort_values(column,ascending=True)[['country',column]].head(5)
    plt.figure(figsize = (7,3))
    plt.title('Bottom 5 countries in : '+ column)
    sns.barplot(data=bot_10, x="country", y=column);
```
